### Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color='green')
plt.title('График f(t)', fontsize=16, family="serif",)
plt.xlabel('Значения t', fontsize=12, family="serif")
plt.ylabel('Значения f', fontsize=12, family="serif")
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

### Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x**2
y2 = 2*x + 0.5
y3 = -3*x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)

ax1.plot(x, y1)
ax1.set_title('График y1', fontsize=14, family="serif")
ax1.set_xlim([-5, 5])

ax2.plot(x, y2)
ax2.set_title('График y2', fontsize=14, family="serif")

ax3.plot(x, y3)
ax3.set_title('График y3', fontsize=14, family="serif")

ax4.plot(x, y4)
ax4.set_title('График y4', fontsize=14, family="serif")
```

### Задание 4

```python
import pandas as pd
plt.style.use('fivethirtyeight')
```

```python
creditcard = 'creditcard.csv'
```

```python
df = pd.read_csv(creditcard, sep=',')
df.head(4)
```

```python
class_list = df['Class'].value_counts()
print(class_list)
```

```python
class_list.plot(kind='bar')
plt.show()
```

```python
class_list.plot(kind='bar', logy=True)
plt.show()
```

```python
class0 = df.loc[df['Class'] == 0, ['V1']]
class1 = df.loc[df['Class'] == 1, ['V1']]
```

```python
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.xlabel('V1', fontsize=16)
plt.legend()
plt.show()
```
